```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading csvs
bst = pd.read_csv("BST-Time.csv")
hashmap = pd.read_csv("hashmap-Time.csv")
```

Creating year vs run time plots 

```python
bst.plot.scatter(x = "Year", y = "t")
plt.savefig("bst_year_time.png")
```

```python
hashmap.plot.scatter(x = "Year", y = "t")
plt.savefig("hashmap_year_time.png")
```

Total number of words vs run time 

```python
bst.plot.scatter(x = "Total_Word_Count", y = "t")
plt.savefig("bst_total_time.png")
```

```python
hashmap.plot.scatter(x = "Total_Word_Count", y = "t")
plt.savefig("hashmap_total_time.png")
```

Unique word count vs time

```python
bst.plot.scatter(x = "Unique_Word_Count", y = "t")
plt.savefig("bst_unique_time.png")
```

```python
hashmap.plot.scatter(x = "Unique_Word_Count", y = "t")
plt.savefig("hashmap_unique_time.png")
```

Looking at unique impacts to average run time

```python
# Hashmap, looking at how collisions impact run time
hashmap.plot.scatter(x = "Collisions", y = "t")
plt.savefig("hashmap_collisions_time.png")
```

```python
# BST map, looking at how depth impacts run time
bst.plot.scatter(x = "Tree_Depth", y = "t")
plt.savefig("bst_depth_time.png")
```
